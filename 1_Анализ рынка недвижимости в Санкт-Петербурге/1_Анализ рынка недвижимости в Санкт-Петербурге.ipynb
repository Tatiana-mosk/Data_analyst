Видим следующие проблемы в данных:пропуски в данных, некорректные типы данных, в столбце "locality_name" в слове поселок указана буквы "е" и "ё"(что может отразится на обработке дубликатов и общем выводе). Все указанные проблемы необходимо обратотать и устранить, так как они ощутимо влияют на работу с данными и могут существенно именить итоговые выводы анализа.

## Предобработка данных

```python
#проверим наличие дубликатов
# предварительно дубликатов не видим
print(df.duplicated().sum())
```

```python
#проверим наличие пропусков
df.isna().sum().sort_values()
```

```python
#посмотрим количество пропусков в процентах для каждого столбца
df.isna().sum().sort_values()/df.shape[0] * 100
```

```python
#для начала займемся обработкой столбца "locality_name"
#заменим буквы "ё" на "е" в слове "поселок" и проверим
df['locality_name'] = df['locality_name'].str.replace('ё', 'е', regex=True)
print(df['locality_name'].unique())
```

```python
#в столбце "locality_name" всего 49 пропусков, это 0,2 %, удаляем все эти пропуски,
#так как они будут искажать результаты анализа, их количество незначительно, но данные о местонахождении
# недвижимости являются одними из основных при ответах на итоговые вопросы анализа
# также замена этих пропусков более критична для дальнейшей обработки, нежели их удаление

df = df.dropna(subset=['locality_name'])
df['locality_name'].isna().sum()

```

```python
# пропусков в столбце 'floors_total' незначительное количество, заменим их на значение столбца 'floors'
# с прибавлением числа "1"
# логика данной обработки такова: общее количество этажей не может быть меньше этажа определеннного
#объекта недвижимости, указанного в столбце 'floor', но чтобы при дальнейшей обработке данные
# при категоризации не попадали в "послеледний этаж" и не искажали при этом объективность выводов,
# прибавим единицу

df['floors_total'] = df['floors_total'].fillna(df['floor'] + 1)
df['floors_total'].isna().sum()
```

```python
#пропуски в столбце 'livin_area' заполним медианным значением для прощади методом transform
df['living_area'] = df['living_area'].fillna(df.groupby('rooms')['living_area'].transform('median'))
df['living_area'].isna().sum()
```

```python
#пропуски в столбце 'kitchen_area' заполним разницей между значениями "общая площать квартиры" и "жилая площадь"
# значение ноль в таблице датасета может означать отсутствие кухни в квартире_студии
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] - df['living_area'])
df['kitchen_area'].isna().sum()
```

Относительно пропусков в столбце 'days_exposition' есть сомнения в целесообразности их заполнения, так как данный столбец определяет время от начала до снятия объявления с размещения. На дату анализа датасета объявления, в которых пропущены значения в данном столбце могут находиться в режиме размещения. Для проверки гипотезы посмотрим дату последнего размещения объявления и наличие объявлений которые размещены  длительный срок.

```python
# проверим дату последнего размещения объявления
# прошло более двух лет

df['first_day_exposition'].max()
```

```python
# проверим количество объявлений, размещенных более двух лет 

df[df['days_exposition'] > (365*2)].shape[0]
```

Как видим из предыдущих действий, пропусков в столбце 'days_exposition' всего в датасете 3180 (13,42 %), объявлений, которые были размещены более 2-х лет и сняты с размещения на момент анализа - 750. Таким образом, вполне вероятно, что данные пропуски - это фактическое нахождение объявления на размещении на день проведения анализа данных. Поэтому не будем их удалять или заменять иными значениями. 

```python
#заполним пропуски значением 'False' в столбце 'is_appartment', применяя логику: если есть пропуск 
# в данном столбце, то это не аппартаменты
# проверим результат

df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].isna().sum()
```

```python
#пропуске в столбце 'balcony' могут означать вероятнее всего его отсутствие
# поэтому заполним их значением ноль, проверим результат

df['balcony'] = df['balcony'].fillna(value=0)
df['balcony'].isna().sum()
```

```python
# в столбце 'airports_nearest' заменим пропуски на медианные значения
df['airports_nearest'] = df['airports_nearest'].fillna(df.groupby('floor')['airports_nearest'].transform('median'))
df['airports_nearest'].isna().sum()                                                      
```

```python
# в столбце 'cityCenters_nearest' заменим пропуски на медианные значения

df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna(df.groupby('floor')['cityCenters_nearest'].transform('median'))
df['cityCenters_nearest'].isna().sum()    
```

```python
# в столбце 'parks_nearest' заменим пропуски на медианные значения
df['parks_nearest'] = df['parks_nearest'].fillna(df.groupby('open_plan')['parks_nearest'].transform('median'))
df['parks_nearest'].isna().sum()
```

```python
# в столбце 'ponds_nearest' заменим пропуски на медианные значения
df['ponds_nearest'] = df['ponds_nearest'].fillna(df.groupby('open_plan')['ponds_nearest'].transform('median'))
df['ponds_nearest'].isna().sum()
```

```python
# в столбце 'ceiling_height' заменим пропуски на медианные значения
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('open_plan')['ceiling_height'].transform('median'))
df['ceiling_height'].isna().sum()
```

###  Результат обработки пропусков.
Последовательно обработаны пропуски во всех столбцах датасета.
Для некоторых пропущенных значений применена логичная замена. Например, если не указано число балконов — скорее всего, их нет. Такие пропуски замены на 0.
Для других типов данных нет подходящего значения на замену.Например, в столбце "сколько дней было размещено объявление (от публикации до снятия)". В этом случае оставлены эти значения пустыми. 
Возможные причины наличия пропусков в датасете, как было описано выше при выполнении кодов, это фактическое отсутствие того или иного признака объекта недвижимости относительно предлагаемых для заполнения характеристик объекта. Данные предположения учтены при обработке пропусков. Также при заполнении объявлений, граждане могли не обладать информацией о той или иной характеристике объекта, что приводило к пропуску заполнения значений. Данная причина тоже учтена при обработке и где это было логично, заменена на медианные значения.







```python
# изменим в столбце 'days_exposition' тип данных float64 на int64 для удобства дальнейшей обработки данных
# так как количество дней лучше измерять в целых числах, что более логично
# проверим результат

df['days_exposition'] = df['days_exposition'].round().astype('Int64')

df.info()
```

```python
# Создадим функцию, которая выводит describe() колонки df[param] и строит для неё гистограмму с заголовком title
# параметр m2 указала 75% умноженный на 3 (решение принято исходя из попыток построить гистограммы с m2=dx['max']),
#так как в описательной статистике увидим максимальное значение, а на гистограмме
#удобнее изучать не максимальный показатель, для более детального отображения параметров,
#так как при указании максимума гистограмма трудно читаема и малоинформативна
def automat_hist_describe(df, param, title):
    dx = df[param].describe()
    print(dx)
    m1 = dx['min']
    m2 = dx['75%']*3
    df[param].plot(kind='hist', bins=30, range=(m1, m2), grid=True, title=title, figsize=(10,5), legend=True)
```

```python
# изучим общую площадь, построим гистограмму

automat_hist_describe(df, 'total_area', 'площадь, кв.м')
```

```python
# рассмотрим зависимость цены квартиры в центре СПб от удалённости от центра:
plt.figure(figsize=(15,7))
plt.title('Зависимость цены от удалённости от центра (весь СПб и центр СПб)')
plt.xlabel('Количество км от центра')
plt.ylabel('Цена')
plt.plot(spb_df.pivot_table(index='cityCenters_nearest_km', values='price_meter', aggfunc='median'), 'o', alpha=0.4, label='весь СПб')
plt.plot(center_spb_df.pivot_table(index='cityCenters_nearest_km', values='price_meter', aggfunc='median'), 'r*', alpha=1, label='центр СПб')
plt.legend(loc="upper left")
plt.show()

print('Коэффициент корреляции Пирсона (центр СПб):', center_spb_df['price_meter'].corr(center_spb_df['cityCenters_nearest_km']))
print('Коэффициент корреляции Пирсона (весь СПб):', spb_df['price_meter'].corr(spb_df['cityCenters_nearest_km']))
```

**Вывод**: Чем ближе к центру квартиры, тем выше их стоимость за 1 кв.м., причем зависимость цены квартир по всему городу более зависима от приближения к центру Санкт-Петербурга. Это связано с тем, что мы выделили определенный сегмент в 8 км в переменную центра города, и соответственно, эти квартиры уже находятся в самом центре и отдаленности практически нет (только ближе к границе 8 км зоны).

```python
# рассмотрим зависимость цены квартиры в центре города от даты публикации объявления
def price_relation_data(param, title, xlabel):
    plt.figure(figsize=(15,7))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Цена')
    plt.plot(spb_df.pivot_table(index=param, values='price_meter', aggfunc='median'), 'o', linestyle='solid', label='весь СПб')
    plt.plot(center_spb_df.pivot_table(index=param, values='price_meter', aggfunc='median'), 'r*', linestyle='solid', label='центр СПб')
    plt.legend(loc="upper left")
    plt.show()
    
# День недели:
price_relation_data('week_day_exposition', 'Зависимость цены от дня недели публикации объявления (весь СПб и центр СПб)', 'День недели')

# Месяц:
price_relation_data('month_exposition', 'Зависимость цены от месяца публикации объявления (весь СПб и центр СПб)', 'Месяц')

# Год:
price_relation_data('year_exposition', 'Зависимость цены от года публикации объявления (весь СПб и центр СПб)', 'Год')
```

**Вывод**: Цена за 1 кв.м. меняется в зависимости от дня недели, месяца и года размещения объявления как по всему Санкт-Петербургу, так и отдельно в его центре. Но есть некоторые отличия. Так зависимость цены от дня недели размещения в центре города выше в субботу(повышается). Скорее всего это связано с наличием наплыва туристов, приездом на выходные дни покупателей, желающих купить квартиру именно в центре, так как растпространено общее мнение, что квартира в центре - это лучший вариант. По всему городу цена квартир практически не зависит от дня размещения обьявления. Относительно повышения цены в центре города в марте и августе, могу предположить, что август - это месяц перед началом учебы в высших учебных заведениях и идет наплав покупателей, повышается спрос и соответственно цена, причем в центре города она имеет большее повышение. Относительно повышения цен в марте месяце могу предположить, что это связано с потеплением и также увеличением спроса, причем в центре города эта зависимость сильнее. Интересна зависимость цены от года размещения объявления, с 2014 года она понижалась незначительно как в центре Санкт-Петербурга, так и по всему городу, когда как по всему объему данных, цена на квартиры сильно снижалась с середины 2014 года (вероятно это может быть связано с началом запуска проекта Я.Недвижимость в конце 2014 года и изначально были выставлены на продажу квартиры с высокой ценой, которые не были востребованы у покупателя ранее именно из-за такой стоимости). Что касается роста цен на квартиры в центре Санкт-Петербурга и по всему городу, то с 2015 цена постепенно растет без сильных скачков. Рост цен связываю с ростом инфляции.
